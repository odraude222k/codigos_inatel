```python
#escolhendo pinguim favorito

from PIL import Image  # Importa a biblioteca PIL para manipular imagens
import numpy as np
import matplotlib.pyplot as plt

def aplicar_filtro_vermelho(imagem):
   # Converte a imagem para tons de cinza para análise
    imagem_cinza = np.mean(imagem, axis=2).astype(np.uint8)  # Média das cores para criar a imagem em escala de cinza
    
   # Cria uma máscara para pixels escuros (que são possíveis pinguins)
    mascara = imagem_cinza < 200  # Definir os pinguins como pixels escuros (intensidade abaixo de 200)

    # Modificar os pixels que fazem parte do pinguim para vermelhos
    imagem[mascara] = [255, 0, 0]  # Troca para a cor vermelha (255, 0, 0)
    
    return imagem



# Carregar as imagens 
img1 = np.array(Image.open(r'C:\Users\perei\OneDrive\Área de Trabalho\faculdade 2025.1\CG-lab\Projeto 1 2025\Projeto 1\P1.png').convert('RGB'))
img2 = np.array(Image.open(r'C:\Users\perei\OneDrive\Área de Trabalho\faculdade 2025.1\CG-lab\Projeto 1 2025\Projeto 1\P2.png').convert('RGB'))
img3 = np.array(Image.open(r'C:\Users\perei\OneDrive\Área de Trabalho\faculdade 2025.1\CG-lab\Projeto 1 2025\Projeto 1\P3.jpg').convert('RGB'))
img4 = np.array(Image.open(r'C:\Users\perei\OneDrive\Área de Trabalho\faculdade 2025.1\CG-lab\Projeto 1 2025\Projeto 1\P4.png').convert('RGB'))

# Obter largura e altura mínimas para padronizar o tamanho
largura_min = min(img1.shape[1], img2.shape[1], img3.shape[1], img4.shape[1])
altura_min = min(img1.shape[0], img2.shape[0], img3.shape[0], img4.shape[0])

# Redimensionar as imagens para o menor tamanho encontrado
img1 = np.array(Image.fromarray(img1).resize((largura_min, altura_min), Image.LANCZOS))
img2 = np.array(Image.fromarray(img2).resize((largura_min, altura_min), Image.LANCZOS))
img3 = np.array(Image.fromarray(img3).resize((largura_min, altura_min), Image.LANCZOS))
img4 = np.array(Image.fromarray(img4).resize((largura_min, altura_min), Image.LANCZOS))

#--------------------- começa a alteração para reflexão ----------------------------------

(l, c, p) = img1.shape #dimensões
(l, c, p) = img2.shape #dimensões
(l, c, p) = img3.shape #dimensões
(l, c, p) = img4.shape #dimensões

img1_refl = np.zeros(shape=img1.shape, dtype=np.uint8) # Cria um novo array de zeros com o mesmo shape da imagem original. Esse array será usado para armazenar a imagem refletida.
img2_refl = np.zeros(shape=img2.shape, dtype=np.uint8) 
img3_refl = np.zeros(shape=img3.shape, dtype=np.uint8)
img4_refl = np.zeros(shape=img4.shape, dtype=np.uint8)

def aplicar_reflexao(img: np.array, img_refl):
    for i in range(l):
         for j in range(c):
            new_x = j
            new_y = l - 1 - i 
            img_refl[new_y, new_x] = img[i, j]

    return img_refl

img1 = aplicar_reflexao(img1,img1_refl)
img2 = aplicar_reflexao(img2,img2_refl)
img3 = aplicar_reflexao(img3,img3_refl)
img4 = aplicar_reflexao(img4,img4_refl)


img1 = aplicar_filtro_vermelho(img1)
img2 = aplicar_filtro_vermelho(img2)
img3 = aplicar_filtro_vermelho(img3)
img4 = aplicar_filtro_vermelho(img4)

# Criar a imagem final como um grid 2x2
img_final = np.zeros((altura_min * 2, largura_min * 2, 3), dtype=np.uint8)

# Posicionar as imagens no grid
img_final[:altura_min, :largura_min] = img1  # Superior esquerdo
img_final[:altura_min, largura_min:] = img2  # Superior direito
img_final[altura_min:, :largura_min] = img3  # Inferior esquerdo
img_final[altura_min:, largura_min:] = img4  # Inferior direito

top, bottom = 0, 230
left, right = 0, 200

mask = np.zeros(shape=img_final.shape, dtype=np.uint8)
mask[top:bottom, left:right, :] = 255  # A área onde o pinguim está fica com valor 255

img_masked = img_final & mask

def histogram_gray(img):
    (l, c) = img.shape
    # Histograma com 1 linha e 256 colunas, preenchido com 0.
    histogram = np.zeros(256, dtype=np.int32)

    for i in range(l):
        for j in range(c):
            # Normalizar intensidade do pixel 
            intensity = img[i, j]
            # Incrementar de 1 a quantidade de pixels no histograma,
            # utilizando a intensidade como índice.
            histogram[intensity] += 1
    return histogram

def show_histogram_gray(img):
    histogram = histogram_gray(img)
    plt.figure(figsize=(16, 16))
    plt.subplot(2, 1, 1)
    # vmin e vmax não permitem que o matplotlib faça normalização das cores.
    plt.imshow(img, cmap='gray', vmin=0, vmax=255)
    plt.subplot(2, 1, 2)
    plt.bar(range(256), histogram)

def show_histogram(img):
    histogram_r = histogram_gray(img[:, :, 0])
    histogram_g = histogram_gray(img[:, :, 1])
    histogram_b = histogram_gray(img[:, :, 2])
    
    plt.figure(figsize=(16, 16))
    plt.subplot(2, 1, 1)
    plt.imshow(img, cmap='gray', vmin=0, vmax=255)
    plt.subplot(2, 1, 2)
    plt.bar(range(256), histogram_r, color='red')
    plt.bar(range(256), histogram_g, color='green')
    plt.bar(range(256), histogram_b, color='blue')


# Transformando a imagem em tons de cinza numa imagem RGB e mantendo as cores
img_com_th = np.stack([img_masked_gray] * 3, axis=-1)  # (H, W) -> (H, W, 3)

# Definir o threshold
threshold = 100

# Criar uma máscara para os pixels abaixo do threshold
mask = img_masked_gray < threshold  # A máscara continua sendo 2D

# Aplicar a cor somente nos locais abaixo do threshold
img_com_th[mask] = [255 , 255, 0]  # Agora a imagem é RGB e pode receber 3 valores

# Exibir a imagem resultante
plt.imshow(img_com_th)
plt.show()




```
